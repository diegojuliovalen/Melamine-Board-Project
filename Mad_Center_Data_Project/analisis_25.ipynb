TOTAL SALES PER STORE

```python
import pandas as pd
import numpy as np

mad_df = pd.read_csv(
    "mad_center_data.csv", delimiter=";", parse_dates=["FECHA"], index_col="FECHA")

mad_df = mad_df[mad_df.index.year == 2025]
mad_df.groupby("SUCURSAL")["TOTAL"].sum().sort_values(ascending=False)
```

TYPE OF PRODUCT RANKING

```python
mad_df["FAMILIA"].value_counts(dropna=False).head(10)
```

TOP VENDORS IN THE COMPANY

```python
mad_df.groupby("VENDEDOR")["TOTAL"].sum().sort_values(ascending=False)
```

TOP TABLETOP PRODUCTS PER STORE

```python
mad_df["CANT."] = pd.to_numeric(mad_df["CANT."], errors="coerce")

melaminas = mad_df[
    mad_df["Sustrato"].isin(["MADERADO", "SYNCHRONICO"]) |
    (mad_df["FAMILIA"] == "TABLERO")
]


top10 = (
    melaminas
    .groupby(["SUCURSAL", "DESCRIPCIÓN"])["CANT."]
    .sum()
    .sort_values(ascending=False)
    .groupby(level=0)
    .head(20)
)

for sucursal, datos in top10.groupby(level=0):
    print(f"\n===== {sucursal} =====")
    print(datos.reset_index(level=0, drop=True))
```

TOP TABLETOP PRODUCTS PER STORE ONLY 15MM

```python
mad_df["CANT."] = pd.to_numeric(mad_df["CANT."], errors="coerce")

melaminas = mad_df[
    mad_df["Sustrato"].isin(["MADERADO", "SYNCHRONICO"]) |
    (mad_df["FAMILIA"] == "TABLERO")
]

melaminas_15mm = melaminas[melaminas["Espesor / Medida"] == "15MM"]

top20_15mm = (
    melaminas_15mm
    .groupby(["SUCURSAL", "DESCRIPCIÓN"])["CANT."]
    .sum()
    .sort_values(ascending=False)
    .groupby(level=0)
    .head(20)
)

for sucursal, datos in top20_15mm.groupby(level=0):
    print(f"\n===== {sucursal} =====")
    print(datos.reset_index(level=0, drop=True))
```

```python
mad_df["CANT."] = pd.to_numeric(mad_df["CANT."], errors="coerce")

melaminas = mad_df[
    mad_df["Sustrato"].isin(["MADERADO", "SYNCHRONICO"]) |
    (mad_df["FAMILIA"] == "TABLERO")
]
melaminas_18mm = melaminas[melaminas["Espesor / Medida"] == "18MM"]

top20_18mm = (
    melaminas_18mm
    .groupby(["SUCURSAL", "DESCRIPCIÓN"])["CANT."]
    .sum()
    .sort_values(ascending=False)
    .groupby(level=0)
    .head(20)
)

for sucursal, datos in top20_18mm.groupby(level=0):
    print(f"\n===== {sucursal} (18MM) =====")
    print(datos.reset_index(level=0, drop=True))
```

TOP TABLETOP PRODUCTS PER STORE ALL OTHERS

```python
mad_df["CANT."] = pd.to_numeric(mad_df["CANT."], errors="coerce")

melaminas = mad_df[
    mad_df["Sustrato"].isin(["MADERADO", "SYNCHRONICO"]) |
    (mad_df["FAMILIA"] == "TABLERO")
]
otros_espesores = melaminas[~melaminas["Espesor / Medida"].isin(["15MM", "18MM"])]

top20_otros = (
    otros_espesores
    .groupby(["SUCURSAL", "DESCRIPCIÓN"])["CANT."]
    .sum()
    .sort_values(ascending=False)
    .groupby(level=0)
    .head(20)
)

for sucursal, datos in top20_otros.groupby(level=0):
    print(f"\n===== {sucursal} (OTROS ESPESORES) =====")
    print(datos.reset_index(level=0, drop=True))
```

GLOBAL RANKING OF 15MM TABLETOPS 2025 

```python
mad_df["CANT."] = pd.to_numeric(mad_df["CANT."], errors="coerce")

melaminas = mad_df[
    mad_df["Sustrato"].isin(["MADERADO", "SYNCHRONICO"]) |
    (mad_df["FAMILIA"] == "TABLERO")
]
melaminas_15mm = melaminas[melaminas["Espesor / Medida"] == "15MM"]

ranking_15mm = (
    melaminas_15mm
    .groupby("DESCRIPCIÓN")["CANT."]
    .sum()
    .sort_values(ascending=False)
)

ranking_15mm.head(20)
```

GLOBAL RANKING OF 18MM TABLETOPS 2025

```python
melaminas_18mm = melaminas[melaminas["Espesor / Medida"] == "18MM"]

ranking_18mm = (
    melaminas_18mm
    .groupby("DESCRIPCIÓN")["CANT."]
    .sum()
    .sort_values(ascending=False)
)

ranking_18mm.head(20)
```

GLOBAL RANKING OF SMALL TABLETOPS 2025

```python
otros_espesores = melaminas[~melaminas["Espesor / Medida"].isin(["15MM", "18MM"])]

ranking_otros = (
    otros_espesores
    .groupby("DESCRIPCIÓN")["CANT."]
    .sum()
    .sort_values(ascending=False)
)

ranking_otros.head(20)
```

GLOBAL AVERAGE PRICE PER STORE 2025

```python
data = melaminas.copy() 

data["THICKNESS_CAT"] = np.where(
    data["Espesor / Medida"].isin(["15MM", "18MM"]),
    data["Espesor / Medida"].str.upper(),
    "OTHER"
)

pivot_branch = (
    data
    .groupby(["SUCURSAL", "THICKNESS_CAT"])["PRECIO"]
    .mean()
    .round(2)
    .unstack(fill_value=np.nan)
    .reset_index()
    
)

print("Average price per MM:")
print(pivot_branch.head(10))


overall = (
    data
    .groupby("THICKNESS_CAT")["PRECIO"]
    .mean()
    .round(2)
    .reset_index()
    .rename(columns={"PRECIO": "Average Price per MM"})
)

print("\nOverall average price 2025:")
print(overall)
```
